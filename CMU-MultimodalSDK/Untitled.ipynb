```python
import torch
from torch.utils.data import Dataset

class CMUMOSEIDataset(Dataset):
    def __init__(self, data, modalities=['glove_vectors', 'COVAREP', 'OpenFace_2'], max_seq_len=50):
        """
        Args:
            data: mmdataset  align 
            modalities: 
            max_seq_len: 0
        """
        self.data = data
        self.modalities = modalities
        self.max_seq_len = max_seq_len
        
        # 
        self.samples = []
        for video_id in list(data['All Labels'].keys()):
            for segment in data['All Labels'][video_id].keys():
                self.samples.append((video_id, segment))
        
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        video_id, segment = self.samples[idx]
        
        # 
        features = []
        for modality in self.modalities:
            feature = self.data[modality][video_id][segment]['features']  # (T, D)
            feature = torch.tensor(feature, dtype=torch.float)  #  tensor
            features.append(feature)
        
        # 
        label = self.data['All Labels'][video_id][segment]['features'][0][0]  # (1,1)scalar
        
        #  (padding)
        padded_features = []
        for feature in features:
            feature = self._pad_or_truncate(feature)
            padded_features.append(feature)
        
        # 
        multimodal_feature = torch.cat(padded_features, dim=-1)  # (max_seq_len, total_dim)
        
        return multimodal_feature, torch.tensor(label, dtype=torch.float)
    
    def _pad_or_truncate(self, feature):
        """padmax_seq_len"""
        T, D = feature.shape
        if T >= self.max_seq_len:
            feature = feature[:self.max_seq_len, :]
        else:
            pad = torch.zeros((self.max_seq_len - T, D))
            feature = torch.cat([feature, pad], dim=0)
        return feature
```
