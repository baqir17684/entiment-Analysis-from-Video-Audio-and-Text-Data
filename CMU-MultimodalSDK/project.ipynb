```python
import torch
from torch.utils.data import Dataset

class CMUMOSEIDataset(Dataset):
    def __init__(self, data, modalities=['glove_vectors', 'COVAREP', 'OpenFace_2'], max_seq_len=50):
        """
        Args:
            data: mmdataset  align 
            modalities:  ['glove_vectors', 'COVAREP', 'OpenFace_2']
            max_seq_len: 0
        """
        self.data = data
        self.modalities = modalities
        self.max_seq_len = max_seq_len
        
        # 
        self.samples = []
        for video_id in list(data['All Labels'].keys()):
            for segment in data['All Labels'][video_id].keys():
                self.samples.append((video_id, segment))
        
    def __len__(self):
        return len(self.samples)
    
    def __getitem__(self, idx):
        video_id, segment = self.samples[idx]
        
        # 
        label = self.data['All Labels'][video_id][segment]['features'][0][0]  # shape (1,1) scalar
        label = torch.tensor(label, dtype=torch.float)
        
        #  padding 
        text_feature  = self._get_and_pad_modality('glove_vectors', video_id, segment)
        audio_feature = self._get_and_pad_modality('COVAREP',      video_id, segment)
        visual_feature= self._get_and_pad_modality('OpenFace_2',   video_id, segment)
        
        #  [max_seq_len, 300], [max_seq_len, 74], [max_seq_len, 35],  label
        return text_feature, audio_feature, visual_feature, label
    
    def _get_and_pad_modality(self, modality, video_id, segment):
        """ pad/truncate."""
        feature = self.data[modality][video_id][segment]['features']  # (T, D)
        feature = torch.tensor(feature, dtype=torch.float)
        return self._pad_or_truncate(feature)
    
    def _pad_or_truncate(self, feature):
        """ pad/truncate  max_seq_len."""
        T, D = feature.shape
        if T >= self.max_seq_len:
            feature = feature[:self.max_seq_len, :]
        else:
            pad = torch.zeros((self.max_seq_len - T, D))
            feature = torch.cat([feature, pad], dim=0)
        return feature
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from project1 import CMUMOSEIDataset                #  Dataset 
from test_mosi import MFN                                #  MFN 
import pickle
from mmsdk import mmdatasdk

cmumosei_highlevel = mmdatasdk.mmdataset("cmumosei_highlevel.pkl")
# ==========  ==========
batch_size = 32
num_epochs = 10
learning_rate = 1e-3
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ==========  ==========
dataset = CMUMOSEIDataset(cmumosei_highlevel)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# ==========  ==========
config = {
    "input_dims": [300, 74, 35],
    "h_dims": [128, 32, 32],
    "memsize": 64,
    "windowsize": 2
}
NN1Config = {"shapes": 128, "drop": 0.3}
NN2Config = {"shapes": 64, "drop": 0.3}
gamma1Config = {"shapes": 64, "drop": 0.3}
gamma2Config = {"shapes": 64, "drop": 0.3}
outConfig = {"shapes": 32, "drop": 0.3}

model = MFN(config, NN1Config, NN2Config, gamma1Config, gamma2Config, outConfig).to(device)

# ==========  ==========
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# ==========  ==========
print("  MFN ...")
for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0

    for x, y in dataloader:
        # x: [B, T, D] -> [T, B, D]
        x = x.permute(1, 0, 2).to(device)
        y = y.to(device).float()  # [B]

        optimizer.zero_grad()
        output = model(x)  # [B, 1]
        loss = criterion(output.squeeze(), y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(dataloader)
    print(f"[Epoch {epoch+1}/{num_epochs}] Loss: {avg_loss:.4f}")

# ==========  ==========
torch.save(model.state_dict(), "mfn_model.pt")
print(" mfn_model.pt")
```
